```python
1.야구게임
중복되지 않는 3자리 숫자를 컴퓨터가 정하면 그 숫자를 맞출 때 까지 반복한다.(0은 제외)
숫자와 자리 위치까지 일치하면 스트라이크
숫자는 있는데 자리는 맞지 않다면 볼
스트라이크가 3개가 되면 맞춘 것이므로 멈춘다.
몇번 만에 맞춘 것인지 횟수도 체크한다.

컴퓨터 246 입력 128 1볼
컴퓨터 246 입력 279 1스트라이크
```

```python
import random
number = str(random.randint(1,999)).zfill(3)
print(number)
count = 0

while True:
    guess = input('입력 :')
    count += 1
    if guess == number:
        print('정답입니다')
        break
    else:
        strike = 0
        ball = 0
        for i in range(3):
            if guess[i] == number[i]:
                strike += 1
            elif guess[i] in number:
                ball+=1
        print("스트라이크:{} 볼:{}".format(strike,ball))
print("%d번 만에 맞추셨습니다." % (count))
```

```python
2. 사칙연산게임
컴퓨터가 사칙연산 문제를 내면 답을 맞추는 게임
컴퓨터는 두 개의 숫자와 연산자를 무작위로 선택한다.
숫자값은 1~50 범위에서 선택
사칙연산만 가능, 나누기는 결과값을 정수로 변경 후 처리.

컴퓨터 2 + 4
       5 - 8
       3 * 4
       8 / 3     - > 답은 2로 처리
```

```python
import random
import time

def make_question():
    operand1 = str(random.randint(1, 50))
    operand2 = str(random.randint(1, 50))
    operator = random.randint(1,4)


    if (operator == 1):
        q = operand1 + ' + ' + operand2
    if (operator == 2):
        q = operand1 + ' - ' + operand2
    if (operator == 3):
        q = operand1 + ' * '+ operand2
    if (operator == 4):
        q = operand1 + ' / '+ operand2

    return q

correct = 0
wrong = 0

for x in range(1, 6):
    question = make_question()
    print(question)
    start = time.time()
    user = float(input("= "))
    end = time.time()
    answer = round(eval(question), 2)
    
    if (user == answer):
        print("정답!")
        correct = correct+1
    else :
        print("오답!")
        print(answer)
        wrong = wrong+1
```
